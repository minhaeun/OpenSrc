```python
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import pandas as pd
##글꼴
rc('font', family = 'AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('202012_202012_연령별인구현황_월간.xlsx')
standard = '서울특별시 광진구 '
df.set_index('행정기관', inplace = True)
df = df.T
df2 = df.sub(df[standard], axis = 0)
df3 = df2**2
df3.loc['총합'] = df3.sum()

df3.drop(standard, axis = 1, inplace = True)
different_value = df3.loc['총합'].max()
similar_value = df3.loc['총합'].min()
different = df3.loc['총합'].idxmax(axis = 1)
similar = df3.loc['총합'].idxmin(axis = 1)

df[standard].plot(color = 'g')
df[different].plot(label = different + '(' + str(different_value/101) + ')'  , color = 'r')
df[similar].plot(label = similar + '(' + str(similar_value/101) + ')', color = 'b')
plt.title('광진구와 인구 구조가 가장 비슷한 지역과 가장 다른 지역')
plt.style.use('ggplot')
plt.legend(loc=(1.1,0.5))
plt.show()
```
