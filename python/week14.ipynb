```python
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import pandas as pd
##
rc('font', family = 'AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('202012_202012__.xlsx')
standard = '  '
df.set_index('', inplace = True)
df = df.T
df2 = df.sub(df[standard], axis = 0)
df3 = df2**2
df3.loc[''] = df3.sum()

df3.drop(standard, axis = 1, inplace = True)
different_value = df3.loc[''].max()
similar_value = df3.loc[''].min()
different = df3.loc[''].idxmax(axis = 1)
similar = df3.loc[''].idxmin(axis = 1)

df[standard].plot(color = 'g')
df[different].plot(label = different + '(' + str(different_value/101) + ')'  , color = 'r')
df[similar].plot(label = similar + '(' + str(similar_value/101) + ')', color = 'b')
plt.title('        ')
plt.style.use('ggplot')
plt.legend(loc=(1.1,0.5))
plt.show()
```
