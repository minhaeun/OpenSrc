```python
import csv
import matplotlib.pyplot as plt
file = open('seoul2.csv', 'r', encoding='cp949')
data = csv.reader(file)
next(data)
high = []
low = []
for row in data :
    if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '15':
        high.append(float(row[4]))
        low.append(float(row[3]))

plt.plot(high,'pink')
plt.plot(low,'skyblue')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
file = open('seoul2.csv', 'r', encoding='cp949')
data = csv.reader(file)
next(data)
jan = []
aug = []

for row in data :
    month = row[0].split('-')[1]
    if month == '01':
        jan.append(float(row[-1]))
    elif month == '08':
        aug.append(float(row[-1]))
    else:
        continue
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.legend()
plt.show()
```
